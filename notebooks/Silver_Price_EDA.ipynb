```python
# import relevant packages
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
import pandas_ta as ta
```

```python
# import silver data for the month of june, 2025
ticker = 'SLV'
silver_data_june_25 = yf.download(ticker, start = '2025-06-01', end = '2025-06-30')
```

```python
silver_june_25 = silver_data_june_25.reset_index()
```

```python
silver_june_25
```

```python
x = silver_june_25['Date']
y = silver_june_25['Close']
```

```python
fig = plt.subplots(figsize=(12,5))
plt.plot(x,y)
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Silver Price - USD')

ticks = list(silver_june_25['Date'])
plt.xticks(ticks, rotation =45)

plt.show()
```

```python
sp_500_2020_2025 = yf.download('^SPX', start = '2020-06-01', end = '2025-06-30')
sp_500_2020_2025 =sp_500_2020_2025.reset_index()

x2 = sp_500_2020_2025['Date']
y2 = sp_500_2020_2025['Close'].values.flatten()
```

```python
fig, ax1 = plt.subplots(figsize = (12,5))
ax1.plot(x, y, label = 'silver price')
plt.show()
```

```python
fig,ax1 = plt.subplots(figsize=(12,5))

ax1.plot(x,y, label ='Silver')
ax1.set_ylabel('Silver Price')


ax2 = ax1.twinx()
ax2.plot(x2,y2,label = 'SPX',color = 'Orange')
ax2.set_ylabel('SPX')

fig.legend(loc = 'upper left',bbox_to_anchor = (0.15,0.85))

plt.show()
```
